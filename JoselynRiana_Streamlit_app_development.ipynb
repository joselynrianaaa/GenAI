```python
from rembg import remove
from PIL import Image
import io
import numpy as np
import streamlit as st
import matplotlib.pyplot as plt
```

```python
def remove_background(input_image):
    with io.BytesIO() as output:
        input_image.save(output, format="PNG")
        input_data = output.getvalue()
        result = remove(input_data)  # Perform background removal using rembg
        img_without_bg = Image.open(io.BytesIO(result)).convert("RGBA")
        return img_without_bg
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
for filename in uploaded.keys():
    image = Image.open(io.BytesIO(uploaded[filename]))

    # Display original image
    print("Original Image")
    plt.imshow(image)
    plt.axis('off')
    plt.show()

    # Step 6: Remove the background
    img_without_bg = remove_background(image)

    # Display background removed image
    print("Image with Background Removed")
    plt.imshow(img_without_bg)
    plt.axis('off')
    plt.show()

```
